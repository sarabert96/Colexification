```
import wikipediaapi
```

```

wiki_wiki = wikipediaapi.Wikipedia('en')

page_py = wiki_wiki.page('Python_(programming_language)')
```

```
page_py.text
```

The page is a big string. We have to change its format into a list of lists of words.

```
len(page_py.text)
```

```
sentences = page_py.text.split('.') # very poor splitting
```

```
words=[]
for s in sentences:
  s=s.lower()
  words.append(s.split())
```

```
words
```

Create a model with the wikipedia page, window size 3

```
from gensim.models import FastText
from gensim import utils
```

```
model = FastText(window=3, min_count=1)
model.build_vocab(sentences=words)
total_examples = model.corpus_count
model.train(sentences=words, total_examples=total_examples, epochs=5)
```

```
model["python"]
```

```
sim_score = model.similarity('python','programming')
print(sim_score)
```

```
sim_score = model.similarity('python','natural')
print(sim_score)
```

```
sim_score = model.similarity('python','cinema')
print(sim_score)
```

```
not_matching = model.wv.doesnt_match("python linux programming nature".split())
print(not_matching)
```
